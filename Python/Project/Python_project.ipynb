```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("\nUnique Values:\n", df.nunique())
```

```python
plt.figure(figsize=(6, 5))
df['type'].value_counts().plot(kind='bar', color=['skyblue', 'pink'])
plt.title("Distribution of Content Type")
plt.xlabel("Type of Content")
plt.ylabel("Count")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(8, 5))
plt.barh(top_countries.index, top_countries.values, color='skyblue', edgecolor='black')
plt.title('Top 10 Countries on Netflix', fontsize=14)
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df['genre_main'] = df['listed_in'].apply(lambda x: x.split(',')[0])
top_genres = df['genre_main'].value_counts().head(10)
plt.figure(figsize=(7, 5))
plt.bar(top_genres.index, top_genres.values, color='lightblue', edgecolor='black')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='release_year', hue='type', palette='Set2', order=df['release_year'].value_counts().index[:10])
plt.title("Top 10 Release Years by Type")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False, cmap='YlGnBu')
plt.title("Missing Value Heatmap")
plt.show()
```
